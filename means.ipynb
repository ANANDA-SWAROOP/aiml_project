```python
# https://colab.research.google.com/drive/1jmxEKMck90OPbacS_efEplQeY4CV36xn?usp=sharing

# Install libraries (only needed once in Colab)
# !pip install matplotlib seaborn scikit-learn scipy

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans, AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# -----------------------------
# Load IRIS Dataset
# -----------------------------
iris = load_iris(as_frame=True)
df = iris.frame
print(df.head())

# -----------------------------
# 1. LINEAR REGRESSION
# -----------------------------
# Predict petal length from petal width
X = df[["petal width (cm)"]]
y = df["petal length (cm)"]

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Plot regression line
plt.figure(figsize=(8,5))
plt.scatter(X, y, color="blue", label="Data points")
plt.plot(X, y_pred, color="red", linewidth=2, label="Regression line")
plt.title("Linear Regression on Iris Dataset")
plt.xlabel("Petal Width (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend()
plt.show()

print("Linear Regression Coefficient:", model.coef_[0])
print("Linear Regression Intercept:", model.intercept_)

# -----------------------------
# 2. K-MEANS CLUSTERING
# -----------------------------
X_clust = df[["sepal length (cm)", "sepal width (cm)"]]

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_clust)

plt.figure(figsize=(8,5))
plt.scatter(X_clust["sepal length (cm)"], X_clust["sepal width (cm)"],
            c=labels, cmap="viridis", s=50)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            c="red", marker="X", s=200, label="Centroids")
plt.title("K-Means Clustering (Iris Sepals)")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend()
plt.show()

# -----------------------------
# 3. HIERARCHICAL CLUSTERING
# -----------------------------
linked = linkage(X_clust, method="ward")

plt.figure(figsize=(10,5))
dendrogram(linked, truncate_mode="lastp", p=15)
plt.title("Hierarchical Clustering Dendrogram (Iris Sepals)")
plt.show()

agg_clust = AgglomerativeClustering(n_clusters=3, linkage="ward")
labels_hc = agg_clust.fit_predict(X_clust)

plt.figure(figsize=(8,5))
plt.scatter(X_clust["sepal length (cm)"], X_clust["sepal width (cm)"],
            c=labels_hc, cmap="plasma", s=50)
plt.title("Hierarchical Clustering (Iris Sepals)")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python

# Install libraries (if not already in Colab)
# !pip install matplotlib seaborn scikit-learn scipy

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_regression, make_blobs
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans, AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# -----------------------------
# 1. LINEAR REGRESSION
# -----------------------------
# Generate synthetic regression dataset
X, y = make_regression(n_samples=100, n_features=1, noise=20, random_state=42)

# Fit Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Plot regression line
plt.figure(figsize=(8,5))
plt.scatter(X, y, color="blue", label="Data points")
plt.plot(X, y_pred, color="red", linewidth=2, label="Regression line")
plt.title("Linear Regression Example")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

print("Linear Regression Coefficient:", model.coef_[0])
print("Linear Regression Intercept:", model.intercept_)

# -----------------------------
# 2. K-MEANS CLUSTERING
# -----------------------------
# Create synthetic clustering dataset
X_clust, y_clust = make_blobs(n_samples=200, centers=3, cluster_std=1.2, random_state=42)

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_clust)

# Plot clusters
plt.figure(figsize=(8,5))
plt.scatter(X_clust[:,0], X_clust[:,1], c=labels, cmap="viridis", s=50)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            c="red", marker="X", s=200, label="Centroids")
plt.title("K-Means Clustering Example")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()

# -----------------------------
# 3. HIERARCHICAL CLUSTERING
# -----------------------------
# Use same dataset for hierarchical clustering
linked = linkage(X_clust, method="ward")

# Plot dendrogram
plt.figure(figsize=(10,5))
dendrogram(linked, truncate_mode="lastp", p=15)
plt.title("Hierarchical Clustering Dendrogram")
plt.show()

# Apply Agglomerative Clustering
agg_clust = AgglomerativeClustering(n_clusters=3, linkage="ward")
labels_hc = agg_clust.fit_predict(X_clust)

# Plot clusters
plt.figure(figsize=(8,5))
plt.scatter(X_clust[:,0], X_clust[:,1], c=labels_hc, cmap="plasma", s=50)
plt.title("Hierarchical Clustering Example")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

```

```python
# Install libraries (only needed once in Colab)
# !pip install matplotlib seaborn scikit-learn scipy

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans, AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# -----------------------------
# Load Wine Dataset
# -----------------------------
wine = load_wine(as_frame=True)
df = wine.frame
print(df.head())

# -----------------------------
# 1. LINEAR REGRESSION
# -----------------------------
# Example: Predict "flavanoids" from "alcohol"
X = df[["alcohol"]]
y = df["flavanoids"]

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Plot regression line
plt.figure(figsize=(8,5))
plt.scatter(X, y, color="blue", label="Data points")
plt.plot(X, y_pred, color="red", linewidth=2, label="Regression line")
plt.title("Linear Regression on Wine Dataset")
plt.xlabel("Alcohol")
plt.ylabel("Flavanoids")
plt.legend()
plt.show()

print("Linear Regression Coefficient:", model.coef_[0])
print("Linear Regression Intercept:", model.intercept_)

# -----------------------------
# 2. K-MEANS CLUSTERING
# -----------------------------
# Use "alcohol" and "malic_acid" as clustering features
X_clust = df[["alcohol", "malic_acid"]]

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_clust)

plt.figure(figsize=(8,5))
plt.scatter(X_clust["alcohol"], X_clust["malic_acid"],
            c=labels, cmap="viridis", s=50)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            c="red", marker="X", s=200, label="Centroids")
plt.title("K-Means Clustering (Wine Data)")
plt.xlabel("Alcohol")
plt.ylabel("Malic Acid")
plt.legend()
plt.show()

# -----------------------------
# 3. HIERARCHICAL CLUSTERING
# -----------------------------
linked = linkage(X_clust, method="ward")

plt.figure(figsize=(10,5))
dendrogram(linked, truncate_mode="lastp", p=15)
plt.title("Hierarchical Clustering Dendrogram (Wine Data)")
plt.show()

agg_clust = AgglomerativeClustering(n_clusters=3, linkage="ward")
labels_hc = agg_clust.fit_predict(X_clust)

plt.figure(figsize=(8,5))
plt.scatter(X_clust["alcohol"], X_clust["malic_acid"],
            c=labels_hc, cmap="plasma", s=50)
plt.title("Hierarchical Clustering (Wine Data)")
plt.xlabel("Alcohol")
plt.ylabel("Malic Acid")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_olivetti_faces
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import dendrogram, linkage




# ====================================================
# 2. OLIVETTI FACES DATASET
# ====================================================
faces = fetch_olivetti_faces(shuffle=True, random_state=42)
X_faces = faces.data     # shape = (400, 4096)
y_faces = faces.target   # person ID labels
images = faces.images    # 64x64 images

print("Faces dataset shape:", X_faces.shape)

# -----------------------------
# Show some sample faces
# -----------------------------
fig, axes = plt.subplots(2, 5, figsize=(12, 5))
for i, ax in enumerate(axes.flat):
    ax.imshow(images[i], cmap="gray")
    ax.set_title(f"ID: {y_faces[i]}")
    ax.axis("off")
plt.suptitle("Sample Images from Olivetti Faces Dataset", fontsize=16)
plt.show()

# -----------------------------
# PCA for Visualization
# -----------------------------
pca = PCA(n_components=2)
X_faces_pca = pca.fit_transform(X_faces)

plt.figure(figsize=(8,6))
plt.scatter(X_faces_pca[:,0], X_faces_pca[:,1], c=y_faces, cmap="tab20", s=20)
plt.colorbar(label="Person ID")
plt.title("PCA Projection of Faces Dataset (2D)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()

# -----------------------------
# K-Means Clustering on Faces
# -----------------------------
kmeans_faces = KMeans(n_clusters=40, random_state=42, n_init=10)
labels_faces = kmeans_faces.fit_predict(X_faces)

plt.figure(figsize=(8,6))
plt.scatter(X_faces_pca[:,0], X_faces_pca[:,1], c=labels_faces, cmap="tab20", s=20)
plt.title("K-Means Clustering on Olivetti Faces (PCA Reduced)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()

# Show cluster centroids as "average faces"
fig, axes = plt.subplots(2, 5, figsize=(12,5))
for i, ax in enumerate(axes.flat):
    centroid = kmeans_faces.cluster_centers_[i].reshape(64,64)
    ax.imshow(centroid, cmap="gray")
    ax.set_title(f"Cluster {i}")
    ax.axis("off")
plt.suptitle("K-Means Cluster Centroids (Average Faces)", fontsize=16)
plt.show()

# -----------------------------
# Hierarchical Clustering (on PCA features)
# -----------------------------
linked_faces = linkage(X_faces_pca[:200], method="ward")  # use subset for clarity

plt.figure(figsize=(12,5))
dendrogram(linked_faces, truncate_mode="lastp", p=20)
plt.title("Hierarchical Clustering Dendrogram (Olivetti Faces, PCA)")
plt.show()

agg_faces = AgglomerativeClustering(n_clusters=40, linkage="ward")
labels_faces_hc = agg_faces.fit_predict(X_faces_pca)

plt.figure(figsize=(8,6))
plt.scatter(X_faces_pca[:,0], X_faces_pca[:,1], c=labels_faces_hc, cmap="tab20", s=20)
plt.title("Hierarchical Clustering on Olivetti Faces (PCA Reduced)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

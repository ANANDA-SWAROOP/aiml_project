```python
# https://colab.research.google.com/drive/1jmxEKMck90OPbacS_efEplQeY4CV36xn?usp=sharing

# Install libraries (only needed once in Colab)
# !pip install matplotlib seaborn scikit-learn scipy

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans, AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# -----------------------------
# Load IRIS Dataset
# -----------------------------
iris = load_iris(as_frame=True)
df = iris.frame
print(df.head())

# -----------------------------
# 1. LINEAR REGRESSION
# -----------------------------
# Predict petal length from petal width
X = df[["petal width (cm)"]]
y = df["petal length (cm)"]

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Plot regression line
plt.figure(figsize=(8,5))
plt.scatter(X, y, color="blue", label="Data points")
plt.plot(X, y_pred, color="red", linewidth=2, label="Regression line")
plt.title("Linear Regression on Iris Dataset")
plt.xlabel("Petal Width (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend()
plt.show()

print("Linear Regression Coefficient:", model.coef_[0])
print("Linear Regression Intercept:", model.intercept_)

# -----------------------------
# 2. K-MEANS CLUSTERING
# -----------------------------
X_clust = df[["sepal length (cm)", "sepal width (cm)"]]

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_clust)

plt.figure(figsize=(8,5))
plt.scatter(X_clust["sepal length (cm)"], X_clust["sepal width (cm)"],
            c=labels, cmap="viridis", s=50)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            c="red", marker="X", s=200, label="Centroids")
plt.title("K-Means Clustering (Iris Sepals)")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend()
plt.show()

# -----------------------------
# 3. HIERARCHICAL CLUSTERING
# -----------------------------
linked = linkage(X_clust, method="ward")

plt.figure(figsize=(10,5))
dendrogram(linked, truncate_mode="lastp", p=15)
plt.title("Hierarchical Clustering Dendrogram (Iris Sepals)")
plt.show()

agg_clust = AgglomerativeClustering(n_clusters=3, linkage="ward")
labels_hc = agg_clust.fit_predict(X_clust)

plt.figure(figsize=(8,5))
plt.scatter(X_clust["sepal length (cm)"], X_clust["sepal width (cm)"],
            c=labels_hc, cmap="plasma", s=50)
plt.title("Hierarchical Clustering (Iris Sepals)")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python

# Install libraries (if not already in Colab)
# !pip install matplotlib seaborn scikit-learn scipy

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_regression, make_blobs
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans, AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# -----------------------------
# 1. LINEAR REGRESSION
# -----------------------------
# Generate synthetic regression dataset
X, y = make_regression(n_samples=100, n_features=1, noise=20, random_state=42)

# Fit Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Plot regression line
plt.figure(figsize=(8,5))
plt.scatter(X, y, color="blue", label="Data points")
plt.plot(X, y_pred, color="red", linewidth=2, label="Regression line")
plt.title("Linear Regression Example")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

print("Linear Regression Coefficient:", model.coef_[0])
print("Linear Regression Intercept:", model.intercept_)

# -----------------------------
# 2. K-MEANS CLUSTERING
# -----------------------------
# Create synthetic clustering dataset
X_clust, y_clust = make_blobs(n_samples=200, centers=3, cluster_std=1.2, random_state=42)

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_clust)

# Plot clusters
plt.figure(figsize=(8,5))
plt.scatter(X_clust[:,0], X_clust[:,1], c=labels, cmap="viridis", s=50)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            c="red", marker="X", s=200, label="Centroids")
plt.title("K-Means Clustering Example")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()

# -----------------------------
# 3. HIERARCHICAL CLUSTERING
# -----------------------------
# Use same dataset for hierarchical clustering
linked = linkage(X_clust, method="ward")

# Plot dendrogram
plt.figure(figsize=(10,5))
dendrogram(linked, truncate_mode="lastp", p=15)
plt.title("Hierarchical Clustering Dendrogram")
plt.show()

# Apply Agglomerative Clustering
agg_clust = AgglomerativeClustering(n_clusters=3, linkage="ward")
labels_hc = agg_clust.fit_predict(X_clust)

# Plot clusters
plt.figure(figsize=(8,5))
plt.scatter(X_clust[:,0], X_clust[:,1], c=labels_hc, cmap="plasma", s=50)
plt.title("Hierarchical Clustering Example")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

```

```python


# Install libraries (if not already in Colab)
# !pip install matplotlib seaborn scikit-learn scipy

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_boston, load_wine
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans, AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# -----------------------------
# 1. LINEAR REGRESSION - Boston Housing
# -----------------------------
# Load dataset
boston = load_boston()
df_boston = pd.DataFrame(boston.data, columns=boston.feature_names)
df_boston["MEDV"] = boston.target  # house price

print(df_boston.head())

# Predict house price (MEDV) using RM (average number of rooms)
X = df_boston[["RM"]]
y = df_boston["MEDV"]

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Plot regression line
plt.figure(figsize=(8,5))
plt.scatter(X, y, color="blue", label="Data points")
plt.plot(X, y_pred, color="red", linewidth=2, label="Regression line")
plt.title("Linear Regression: Rooms vs House Price (Boston Housing)")
plt.xlabel("Average Number of Rooms (RM)")
plt.ylabel("House Price (MEDV)")
plt.legend()
plt.show()

print("Linear Regression Coefficient:", model.coef_[0])
print("Linear Regression Intercept:", model.intercept_)

# -----------------------------
# 2. CLUSTERING - Wine Dataset
# -----------------------------
wine = load_wine()
df_wine = pd.DataFrame(wine.data, columns=wine.feature_names)

# Use two features for visualization
X_clust = df_wine[["alcohol", "color_intensity"]]

# ---- KMeans ----
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels_km = kmeans.fit_predict(X_clust)

plt.figure(figsize=(8,5))
plt.scatter(X_clust["alcohol"], X_clust["color_intensity"], c=labels_km, cmap="viridis", s=50)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            c="red", marker="X", s=200, label="Centroids")
plt.title("K-Means Clustering (Wine Data)")
plt.xlabel("Alcohol")
plt.ylabel("Color Intensity")
plt.legend()
plt.show()

# ---- Hierarchical ----
linked = linkage(X_clust, method="ward")

plt.figure(figsize=(10,5))
dendrogram(linked, truncate_mode="lastp", p=15)
plt.title("Hierarchical Clustering Dendrogram (Wine Data)")
plt.show()

agg_clust = AgglomerativeClustering(n_clusters=3, linkage="ward")
labels_hc = agg_clust.fit_predict(X_clust)

plt.figure(figsize=(8,5))
plt.scatter(X_clust["alcohol"], X_clust["color_intensity"], c=labels_hc, cmap="plasma", s=50)
plt.title("Hierarchical Clustering (Wine Data)")
plt.xlabel("Alcohol")
plt.ylabel("Color Intensity")
plt.show()
```
